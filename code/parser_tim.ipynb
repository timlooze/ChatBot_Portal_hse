```python
def urlib_test():
    import urllib.request
    response = urllib.request.urlopen('https://portal.hse.ru')
    html = response.read()
    print(html)
```

```python
matrix = {}
import requests
from bs4 import BeautifulSoup

def bs4_test(url):

    color_mass = set()

    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    urls = []
    for link in soup.find_all('a'):
        try:
            if 'portal.hse' in link.get('href'):
                color_mass.add(link.get('href'))
        except:
            print(link.get('href'), 'mistake')
    matrix[url] = color_mass
    for link in color_mass:
        if link not in matrix.keys():
            print(link)
            bs4_test(link)


bs4_test('https://portal.hse.ru')
print(len(matrix.keys()))
for i in matrix.keys():
    print(i)
```
